```python
import pandas as pd
import csv

combined_data = pd.read_csv('data_archive/data_for_analysis_v2.csv')

for key, value in tournament_dict_3.items():
    country = tournament_dict_3[key]['country']
    
    csv_name = "data_archive/" + country + '_chart_data.csv'
    
    synth_country = "synthetic_" + country
    
    labels = ['year',country,synth_country]
    
    with open(csv_name, "w") as filex:   
        wr = csv.writer(filex, delimiter = ',' , quotechar = '"' )
        wr.writerow(labels)
    
    intercept = float(tournament_dict_3[key]['intercept'])
    
    for yearx in tournament_dict_3[key]['year_list']:
        
        country_value = combined_data[combined_data['year'] == yearx][country].iloc[0]
        
        synth_value = intercept
        
        for keyx,valuex in tournament_dict_3[key]['weights'].items():
            
            control_country = keyx
            
            control_country_weight = valuex
            
            control_country_receipts = combined_data[combined_data['year'] == yearx][control_country].iloc[0]
            
            control_country_partial_receipts = control_country_weight * control_country_receipts
            
            synth_value += control_country_partial_receipts
        
        new_row = [yearx,country_value/1000000000,synth_value/1000000000]
        
        with open(csv_name, "a") as filez:
            wrz = csv.writer(filez, delimiter = ',' , quotechar = '"' )
            wrz.writerow(new_row) 
            
print('done and done')
```
